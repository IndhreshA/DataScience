# Machine Learning

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
churn = df.copy()


labelEncoder = LabelEncoder()

columns = churn.drop(axis=1, columns="Attrition_Flag").select_dtypes(include='object').columns

for col in columns:
    churn[col] = labelEncoder.fit_transform(churn[col])
```

```python
X = churn.drop("Attrition_Flag", axis=1)

transform = StandardScaler()

X = transform.fit_transform(X)
```

```python
y = churn["Attrition_Flag"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Support Vector Machine

```python
parameters = {'kernel':('linear', 'rbf','poly','rbf', 'sigmoid'),
              'C': np.logspace(-3, 3, 5),
              'gamma':np.logspace(-3, 3, 5)}

model = SVC()

grid = GridSearchCV(estimator=model, param_grid=parameters, cv=10)

```

```python
model.fit(X_train, y_train)
```

```python
#model = grid.best_estimator_
```

```python
#print("tuned hpyerparameters :(best parameters) ",grid.best_params_)
#print("accuracy :",grid.best_score_)
```

```python
model.score(X_test, y_test)
```

```python
yhat=model.predict(X_test)
yhat
```

```python
accuracy = accuracy_score(y_test, yhat)
accuracy
```

```python
score = cross_val_score(model,X_train, y_train, cv=10)
np.round(score, 2)
```

```python
np.round(np.mean(score), 2)
```

## Logistic Regression

```python
model = LogisticRegression()

parameters ={'C':[0.01,0.1,1],
             'penalty':['l2'],
             'solver':['lbfgs']}

model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
yhat = model.predict(X_test)
yhat
```

```python
accuracy = accuracy_score(y_test, yhat)
accuracy
```

```python
score = cross_val_score(model,X_train, y_train, cv=10)
np.round(score, 4)
```

```python
np.round(np.mean(score), 4)
```
