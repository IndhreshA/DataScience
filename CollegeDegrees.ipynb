```python
df = pd.read_csv("CollegeDegrees.csv")
```

```python
df['Starting Median Salary'] = df['Starting Median Salary'].str.replace(',', '').str.replace('$', '')
df['Starting Median Salary'] = df['Starting Median Salary'].astype(np.float64).astype(np.int64)

df['Mid-Career Median Salary'] = df['Mid-Career Median Salary'].str.replace(',', '').str.replace('$', '')
df['Mid-Career Median Salary'] = df['Mid-Career Median Salary'].astype(np.float64).astype(np.int64)
```

```python
df[df['Starting Median Salary'] >= 55000]
```

```python
df[df['Mid-Career Median Salary'] >= 85000].head()
```

```python
df.describe()
```

```python
df['Mid-Career 10th Percentile Salary'] = df['Mid-Career 10th Percentile Salary'].str.replace(',', '').str.replace('$', '')
df['Mid-Career 10th Percentile Salary'] = df['Mid-Career 10th Percentile Salary'].astype(np.float64).astype(np.int64)

df['Mid-Career 25th Percentile Salary'] = df['Mid-Career 25th Percentile Salary'].str.replace(',', '').str.replace('$', '')
df['Mid-Career 25th Percentile Salary'] = df['Mid-Career 25th Percentile Salary'].astype(np.float64).astype(np.int64)

df['Mid-Career 75th Percentile Salary'] = df['Mid-Career 75th Percentile Salary'].str.replace(',', '').str.replace('$', '')
df['Mid-Career 75th Percentile Salary'] = df['Mid-Career 75th Percentile Salary'].astype(np.float64).astype(np.int64)

df['Mid-Career 90th Percentile Salary'] = df['Mid-Career 90th Percentile Salary'].str.replace(',', '').str.replace('$', '')
df['Mid-Career 90th Percentile Salary'] = df['Mid-Career 90th Percentile Salary'].astype(np.float64).astype(np.int64)
```

```python
record_over_100k_75th_percentile = np.array(df['Mid-Career 75th Percentile Salary'] >= 100000)
record_over_100k_90th_percentile = np.array(df['Mid-Career 90th Percentile Salary'] >=  100000)
```

```python
df['100k_75th_Percentile'] = record_over_100k_75th_percentile 
df['100k_90th_Percentile'] = record_over_100k_90th_percentile
df.head()
```

# Data Visualization

```python
df.plot(x='Undergraduate Major', y='Percent change from Starting to Mid-Career Salary', kind='hist')
plt.title('Growth from Starting Median Salary')
plt.show()
```

```python
df = df[df['Percent change from Starting to Mid-Career Salary'] > 65.0]
df.plot(x='Undergraduate Major', y='Mid-Career 90th Percentile Salary', kind='bar')
plt.title('Majors with over 65% change from Starting Median Salary')
plt.show()
```

```python
plt.figure(figsize=(26, 13))
df = df[df['Mid-Career 90th Percentile Salary'] >= 170000]
df.plot(y='Mid-Career 90th Percentile Salary', kind='pie', labels=df['Undergraduate Major'], autopct='%1.1f%%', legend=False).set(ylabel=None)
plt.title('Majors with $170,000 salary or higher in the 90th Percentile')
plt.show()
```
